### Operadores Relacionais

```python
# Teste de igualdade numérica (NÃO É =, É ==)

a == b 
```

```python
# Teste de igualdade strings

print(d == e)
print("eu" == "eu")
print("eu" == "Eu")
```

```python
# Teste da diferença numérica
a != b
```

```python
# Teste da diferença strings

print("eu" != "Eu")
```

```python
# Maior
a > b
```

```python
# Menor
a < b
```

```python
# Maior igual
2 >= 2
```

```python
# Menor igual
a <= b
```
