### E como poderamos pegar duas listas separadas e converter em um dicionrio, sem ter que fazer de forma manual?


```python
#Ex: Suponha que tivssemos inicialmente uma lista de nomes e uma lista de notas.
#Esses dados esto ordenados, isto , o primeiro nome na lista 1 corresponde a primeira nota na lista 2 e assim 
#por diante. Como fariamos pra cirar esse dicionrio?
```

```python
lista_nomes = ["Maria", "Joao","Pedro", "Mariana"]
lista_notas = [7.5,9,6.5,8.7]

print(lista_nomes)
print(lista_notas)
```

#### Funo ZIP



```python
#criamos uma lista com tuplas, que sero nossos pares chave:valor no dicionrio
lista_nome_nota = list(zip(lista_nomes, lista_notas)) ; lista_nome_nota
```

#### Operaes entre conjuntos



```python
#Ex: Suponha que vc tem duas listas e quer saber que elementos esto nas duas, isto , esto na primeira lista e na
#segunda lista?
```

```python
pares = [2,4,6,8,10] #elementos pares
multiplos_4 = [4,8,12,16] #elementos mltiplos de 4

pares = set(pares)
multiplos_4 = set(multiplos_4)
```

```python
print(type(pares))
print(type(multiplos_4))
```

```python
pares.
```

```python
# Interseco: Elementos que esto em pares E em multiplos de 4

pares.intersection(multiplos_4)
```

```python
# Unio: elementos que esto em pares OU em mltiplos de 4

pares.union(multiplos_4)
```

```python
# Diferena ( pares - mltiplos de 4): Elementos que esto em pares E NO esto em mltiplos de 4

pares.difference(multiplos_4)
```
