### E como poderíamos pegar duas listas separadas e converter em um dicionário, sem ter que fazer de forma manual?


```python
#Ex: Suponha que tivéssemos inicialmente uma lista de nomes e uma lista de notas.
#Esses dados estão ordenados, isto é, o primeiro nome na lista 1 corresponde a primeira nota na lista 2 e assim 
#por diante. Como fariamos pra cirar esse dicionário?
```

```python
lista_nomes = ["Maria", "Joao","Pedro", "Mariana"]
lista_notas = [7.5,9,6.5,8.7]

print(lista_nomes)
print(lista_notas)
```

#### Função ZIP



```python
#criamos uma lista com tuplas, que serão nossos pares chave:valor no dicionário
lista_nome_nota = list(zip(lista_nomes, lista_notas)) ; lista_nome_nota
```

```python
dicionario_nome_nota = dict(lista_nome_nota)
print(dicionario_nome_nota)
```

## Conjuntos



- Conjuntos não possuem noção de ordem por isso seus, então seus elementos não podem ser acessados com colchetes [ ]
nem podem ser fatiados.

- Conjuntos só aceitam tipos de dados imutáveis como inteiros, floats, tuplas e strings, não aceitando listas,
dicionários e outros conjuntos (set).

#### Criando conjuntos

```python
#Usa chaves também, porém não temos pares e sim elementos
```

```python
primeiro_conjunto = {2,4,6,8}
```

```python
#função set()
#conjuntos não permitem elementos repetidos!

lista_ = [1,2,5,6,6,7,8,8]

print(len(lista_))

print(set(lista_))

print(len(set(lista_)))
```

#### Operações entre conjuntos



```python
#Ex: Suponha que vc tem duas listas e quer saber que elementos estão nas duas, isto é, estão na primeira lista e na
#segunda lista?
```

```python
pares = [2,4,6,8,10] #elementos pares
multiplos_4 = [4,8,12,16] #elementos múltiplos de 4

pares = set(pares)
multiplos_4 = set(multiplos_4)
```

```python
print(type(pares))
print(type(multiplos_4))
```

```python
pares.
```

```python
# Intersecção: Elementos que estão em pares E em multiplos de 4

pares.intersection(multiplos_4)
```

```python
# União: elementos que estão em pares OU em múltiplos de 4

pares.union(multiplos_4)
```

```python
# Diferença ( pares - múltiplos de 4): Elementos que estão em pares E NÃO estão em múltiplos de 4

pares.difference(multiplos_4)
```
