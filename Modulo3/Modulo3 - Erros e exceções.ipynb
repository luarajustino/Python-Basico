```python
#No caso de o erro no estar tratado, nosso cdigo, no momento em que encontra a exceo, ele quebra :(
valor = 1000

parcelas = int(input("Nmero de parcelas: "))

valor_da_parcela = valor/parcelas

print("Ok! Nesse caso o valor de cada parcela :", round(valor_da_parcela,2))
```

```python
#Fazendo o tratamento do erro com try-except:
valor = 1000



try: 
    parcelas = int(input("Nmero de parcelas: "))
    valor_da_parcela = valor/parcelas
    print("Ok! Nesse caso o valor de cada parcela :", round(valor_da_parcela,2))
except:
    print("Algo est errado!")

print("Ser executado porque tratei o erro! :)")
```

```python
# Podemos mostrar qual foi o tipo do erro:
valor = 1000

try: 
    parcelas = int(input("Nmero de parcelas: "))
    valor_da_parcela = valor/parcelas
    print("Ok! Nesse caso o valor de cada parcela :", round(valor_da_parcela,2))
except Exception as erro:
    print("Aconteceu um erro do tipo", erro.__class__, " e a mensagem de erro :",erro)

print("Ser executado porque tratei o erro! :)")
```

```python
# Podemos ainda colocar uma exception para um tipo especfico de erro:

valor = 1000

try: 
    parcelas = int(input("Nmero de parcelas: "))
    valor_da_parcela = valor/parcelas
    print("Ok! Nesse caso o valor de cada parcela :", round(valor_da_parcela,2))
except ZeroDivisionError:
    print("Coloque um nmero diferente de 0")
except ValueError:
    print("Voc s pode por nmeros aqui!")
```
